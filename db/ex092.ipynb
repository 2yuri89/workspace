```python
import pymysql

# 1. 
conn = pymysql.connect(host = "localhost", port = 3306, db = "bookstore",
                       user = "humanda", passwd = "humanda")

# 2.
cursor = conn.cursor()
```

```python
import pandas as pd

winequality_df = pd.read_csv('data-file/winequality-white.csv', sep=";")

winequality_list = winequality_df.values.tolist()
```

```python
# 테이블 만들기
# 자동 증가 번호를 추가해서 pk 설정
sql = """create table winequality_w
         (
            idx integer AUTO_INCREMENT primary key,
            fixed_acidity float,
            volatile_acidity float,
            citric_acid float,
            residual_sugar float,
            chlorides float,
            free_sulfur_dioxide float,
            total_sulfur_dioxide float,
            density float,
            pH float,
            sulphates float,
            alcohol float,
            quality integer
         )"""
cursor.execute(sql)
```

```python
cols = [ col.replace(" ", "_") for col in winequality_df.columns ]
# cols
cols = ','.join(cols)

sql = "insert into winequality_w ({0}) values(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)".format(cols)
```

```python
cursor.executemany(sql, winequality_list)
conn.commit()
```

```python
cursor.close()
conn.close()
```
