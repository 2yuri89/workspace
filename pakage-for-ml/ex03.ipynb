```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.title("Cosine Plot")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
# 각 플롯에 label이 없다면 범례 안 나온다.
plt.legend(loc=2)
# loc = 숫자 > 범례 위치 지정 ppt65쪽
plt.show()
```

```python
# 이건 중요함 > 그래프를 여러개 그려서 한방에 비교하게 함.
# 피규어 하나에 axes를 여러개 둬서 한 코드로 여러 그래프 그리기
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
# plt.subplot(2, 1, 1) > Figure하나에 axes 2개 들어가고, 그중 하나인데 첫 번째다
# subplots도 있는데, 이건 axes를 한방에 만들어. 2행 1열을 한방에 만들어.
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)
plt.show()
```

```python
# 막대그래프
mpl.rc('font', family='malgun gothic')
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = [u'가', u'나', u'다']
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.show()
```

```python
# 파이차트
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold',
'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(sizes, explode=explode, labels=labels,
colors=colors, autopct='%1.1f%%',
shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
