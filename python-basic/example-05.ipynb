##### api활용 - 대기오염 통계현황

* 일자별 미세먼지 (pm10 or pm25) 데이터 조회 및 시각화
* 일정 기간의 미세먼지 데이터를 조회해서 선그래프로 표시
    - 일정 기간 : 올해/작년/재작년 1월~9월

```python
import requests

service_key = "YgC6RL7Xuzrbvz7msHuvRG0JZ+JxR0E6iUI4qrJNk78+5FdBneHQ8wktN9kjkpfJFCUeVtCuvThmYWikuonEuA=="
# 디코딩 인증키

url = 'http://apis.data.go.kr/B552584/ArpltnStatsSvc/getMsrstnAcctoRDyrg'
params ={'serviceKey' : service_key, 'returnType' : 'JSON', 'numOfRows' : '100000', 'pageNo' : '1', 'inqBginDt' : '20220101', 'inqEndDt' : '20241031', 'msrstnName' : '영등포구'}

response = requests.get(url, params=params)
print(response.content)
```

```python
# binary 텍스트를 text 포멧으로 변환하는 작업

dedcoded_response_content = response.content.decode('utf-8')
dedcoded_response_content
```

```python
import json

# json string -> dict 변환
dict_content = json.loads(dedcoded_response_content)
dict_content
```

```python
# 리스트 확인하기

print( dict_content['response'].keys() )
print( dict_content['response']['body'].keys() )
print( type(dict_content['response']['body']['items']) )
print( dict_content['response']['body']['items'] )
```

```python
import pandas as pd

# 판다스 데이터프레임으로 변환
df_content = pd.DataFrame(dict_content['response']['body']['items'])
print(df_content.columns)
```

```python
df_content.shape
df_content.head()
```

```python
df_content["pm10Value"] = df_content['pm10Value'].map(lambda v : float(v))
df_content["pm25Value"] = df_content['pm25Value'].map(lambda v : float(v))

df_content['msurDt'] = pd.to_datetime(df_content['msurDt'])
df_content = df_content.loc[:, ["msurDt", "pm10Value", "pm25Value"]]

df_content.shape
print( df_content )
```

```python
import matplotlib.pyplot as plt

# 맷플롯립 한글 사용
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

plt.figure(figsize=(10, 5))
plt.plot(df_content['msurDt'], df_content['pm10Value'], label = '미세먼지')
plt.plot(df_content['msurDt'], df_content['pm25Value'], label = '초미세먼지')
plt.legend()
plt.show()
```

```python
pm10Max = pd.Series([df_content['pm10Value'].max()] * len(df_content))
pm10Min = pd.Series([df_content['pm10Value'].min()] * len(df_content))
pm25Max = pd.Series([df_content['pm25Value'].max()] * len(df_content))
pm25Min = pd.Series([df_content['pm25Value'].min()] * len(df_content))
print(pm10Max.head())
print(pm10Max.shape)
print(pm10Max.info())
```
