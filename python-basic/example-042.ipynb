```python
import json

text_response = response_body.decode('utf-8')
dict_response = json.loads(text_response)
dict_response
```

```python
print(dict_response['response'].keys())
print(dict_response['response']['body'].keys())
print(dict_response['response']['body']['items'].keys())
print(dict_response['response']['body']['items']['item'])
print(len(dict_response['response']['body']['items']['item']))
```

```python
# 반복하면서 기온(max, min avg) 데이터 뽑기
max_tas = []
min_tas = []
avg_tas = []
dates = []

for item in dict_response['response']['body']['items']['item']:
    # print("최고 기온:", item['maxTa'], "최저 기온:", item['minTa'], "평균 기온:", item['avgTa'])
    max_tas.append( float(item['maxTa']) ) # 문자열로 받은 maxTa를 숫자(실수-float, 정수-int)로 바꾸기
    min_tas.append( float(item['minTa']) ) # 문자열로 받은 minTa를 숫자(실수-float, 정수-int)로 바꾸기
    avg_tas.append( float(item['avgTa']) ) # 문자열로 받은 avgTa를 숫자(실수-float, 정수-int)로 바꾸기
    dates.append(item['tm'])
```

```python
len(min_tas), min_tas[:3]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(dates, max_tas, color = 'red', label = 'max', marker = 'o', alpha = 0.5, linewidth = 2)
plt.plot(dates, min_tas, color = 'blue', label = 'min', marker = 'o', alpha = 0.5, linewidth = 2)
plt.plot(dates, avg_tas, color = 'purple', label = 'avg', ls = '--', marker = '*', alpha = 0.2, linewidth = 1)
plt.xticks(rotation=90)
plt.legend()
plt.show() # 이거 해야 그래프만 나온다.
```

```python
import requests

service_key = "YgC6RL7Xuzrbvz7msHuvRG0JZ+JxR0E6iUI4qrJNk78+5FdBneHQ8wktN9kjkpfJFCUeVtCuvThmYWikuonEuA=="
# Decoding

url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'
params ={'serviceKey' : service_key, 'pageNo' : 1, 'numOfRows' : 40, 'dataType' : 'JSON', 'dataCd' : 'ASOS', 'dateCd' : 'DAY', 'startDt' : '20240921', 'endDt' : '20241021', 'stnIds' : '108' }
# api_key, 1, 40, "JSON", "ASOS", "DAY", '20240921', '20241021', 108

response = requests.get(url, params=params)
print(response.content)
```

```python
import json

text_response = response.content.decode('utf-8')
dict_response = json.loads(text_response)
dict_response
```

```python
import pandas as pd

df_weather = pd.DataFrame(dict_response['response']['body']['items']['item'])
df_weather.head(3)
```

```python
df_weather2 = df_weather[["tm", "minTa", "maxTa", "avgTa"]]
df_weather2.head(3)
```

```python
df_weather2.info() 
# dataFrame을 만들면 반드시 .info()부터 실행해서 null(결측치)찾고, Dtpye보기
# missing value 처리하는거 매우 중요함
# 그 다음 자료형 봐야함
# object는 문자열이라는 뜻
```

```python
print(df_weather2['minTa'].dtype)
print(type(df_weather2['minTa'][0]))
# 컬럼의 데이터는 문자열
```

```python
df_weather2['minTa2'] = df_weather2['minTa'].map(lambda v : float(v))
df_weather2['maxTa2'] = df_weather2['maxTa'].map(lambda v : float(v))
df_weather2['avgTa2'] = df_weather2['avgTa'].map(lambda v : float(v))
# 반복문도 쓸 수 있지만, 이렇게 써야한다. 
# 왜냐면 파이썬에서 반복문쓰면 데이터량이 너무 많아져서, 만들어진 걸 써야함.
```

##### lambda v : float(v) 추가 설명

```python
def f_add(a, b):
	return a + b

f_add(10, 20)

f_add2 = f_add # 함수를 변수에 저장

f_add2(20, 30)

def calc(f, a, b):
	return f(a, b)
	
calc(f_add, 100, 200) # 함수를 전달인자로 사용할 수 있습니다.

f_add3 = lambda a, b: a + b
calc(f_add3, 1000, 2000)

calc(lambda a, b: a + b, 11000, 22000) # 함수 없이 람다처리해서 사용가능

# 함수의 정의를 인라인으로 쓸 수 없다. : 이 코드는 에러남
# calc(def (a, b): return a + b, 1000, 2000)
```
