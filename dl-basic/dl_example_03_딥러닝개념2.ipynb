```python
import tensorflow as tf
from tensorflow import keras as tf_keras
```

```python
from re import X
# 데이터 준비 : keras 제공 내장 데이터셋 사용
(X_train, y_train), (X_test, y_test) = tf_keras.datasets.mnist.load_data()

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
# (60000, 28, 28) (10000, 28, 28) > 이미지 형태로 그대로 가져 옴. cnn이 이 형태 그대로 학습할 수 있는 알고리즘
```

```python
# 입력 데이터 구조 변경
X_train_flatten = X_train.reshape(-1, 28*28)
X_test_flatten = X_test.reshape(-1, 28*28)
print(X_train_flatten.shape, X_test_flatten.shape)
print(X_train_flatten.min(), X_train_flatten.max())

X_train_flatten = X_train_flatten / 255.0     # 입력값이 크면 가중치가 튄다. 그래서 0~1사이로 바꿔줌.
X_test_flatten = X_test_flatten / 255.0
print(X_train_flatten.min(), X_train_flatten.max())
```

```python
# 모델 학습 2-1 - 신경망 사용
model = tf_keras.Sequential([
    tf_keras.layers.Input(shape=(784, )),               # 입력층
    # tf_keras.layers.Dense(512, activation = 'relu'),    # 1번 중간층
    # tf_keras.layers.Dense(512, activation = 'sigmoid'),
        # tf_keras.layers.Dense(512, activation = 'tanh'),
        tf_keras.layers.Dense(512, activation = 'leaky_relu'),
    tf_keras.layers.Dense(256, activation = 'relu'),    # 2번 중간층
    tf_keras.layers.Dense(10, activation = 'softmax')]) # 출력층
```

```python
# 모델 학습 2-3
model.compile(loss='sparse_categorical_crossentropy', # 손실함수
              optimizer='adam',                       # 옵티마이저 # rmsprop 을 옵티마이저로 쓰는 코드도 있다.
              metrics=['accuracy'])                   # 모델평가 방법: 정확도

model.fit(X_train_flatten, y_train, batch_size=256, epochs=10, validation_split=0.2)
```

```python
# 3. 모델 검증 (테스트)

model.evaluate(X_train_flatten, y_train), model.evaluate(X_test_flatten, y_test)
```
